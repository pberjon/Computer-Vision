```python
import math

# Loss function = Cross-entropy: #this compute the gradient by "brut force"
def cross_entropy_bf(y, target,x):
    # s is the prediction of the hidden incoming layer.
    # target is the one-hot vector to compare to x with cross-entropy : sum_j target_j * log(s_j)
    # x is the inout image (only for the gradient)
    # output : cross entropy, gradient of cross-entropy wrt weights
    
    u = np.ones((y.shape[0],1))
    
    # cross-entropy
    ce = -target.T.dot(np.log(y))
    
    # gradient
    g = (-u.T.dot(target)*(y.dot(x.T))+target*x.T)
    #grad =-g.reshape(x.shape[0]*target.shape[0],1)
        
    return ce,g

print(cross_entropy_bf(np.array([0.33333333,0.33333333,0.33333333]), np.array([0.1,0.5,0.25]), np.array([2,2,2])))
```

## Network definition

```python
def NN(x,W):
    
    # x: image, ie 784x1 vector (28x28)
    # W: weights matrices of shape 10x784
    y = W.dot(x)    
    pred = np.array(softmax(y,-np.max(y)))
    return pred,y



 
# main function    
def f(W,x,target):
    
    pred,y = NN(x,W)
    ce,grad = cross_entropy_bf(pred, target,x)
    
    return ce,grad,pred


# function to create Batches of samples
def create_batches(x,bs):
    
    batches = []
    ind = np.arange(x.shape[0])
    random.shuffle(ind)
    nbatch = ind.shape[0]//bs
    rest = ind.shape[0]%bs
    
    for n in range(nbatch):
        batches +=[ind[bs*n:bs*(n+1)]]
    
    # put the remaining elements in a last batch
    if rest !=0:        
        batches += [ind[-rest:]]
        
    return batches
```
